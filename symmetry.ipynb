### Imports

```python
from pymatgen.ext.matproj import MPRester
from pymatgen.io.cif import CifWriter
from pymatgen.symmetry.analyzer import *
from pymatgen.io.cif import *
from pymatgen.core.operations import *
from pymatgen.symmetry.site_symmetries import *
import matplotlib.pyplot as plt
import matplotlib.image as img
```

## Fichier CIF : 

```python
with MPRester("bfWHOpdayyP1LUqo") as m:

    # Structure for material id
    structure = m.get_structure_by_material_id("mp-862947")
```

## Données :

```python
a = SpacegroupAnalyzer(structure).get_symmetry_operations()
```

```python
#On récupère les opérations de symmétrie 
Sym1 = a[1]
Sym2 = a[2]
Sym3 = a[11]

#Coordonnées de 3 atomes dans la maille conventionnelle 
Ba2 = [0.25,0.25,0.25]
Bi = [0,0,0]
Au = [0,0,0.5]

#Applications des opérations de symmétrie
Ba2Sym = Sym1.operate(Ba2)
BiSym = Sym2.operate(Bi)
AuSym = Sym3.operate(Au)


```

## Première symmétrie : 

```python
print("Première opération de symétrie")
print("------------------------------")
print(" ")
print("Atome de départ : Ba2", Ba2 )
print(" ")
print("Opération de symétrie : inversion autour du point (0,0,0) : ")
print(Sym1)
print(" ")
print("Atome d'arrivée : Ba", Ba2Sym)
print(" ")
print("Image :")

image_1=img.imread('data/Ba2.png')
fig, axes = plt.subplots(figsize=(10,10))
axes.get_xaxis().set_visible(False)
axes.get_yaxis().set_visible(False)
axes.imshow(image_1)
plt.show()
```

## Deuxième symmétrie :

```python
print("Deuxième opération de symétrie")
print("------------------------------")
print(" ")
print("Atome de départ : Bi", Bi )
print(" ")
print("Opération de symétrie : Plan miroir : ")
print(Sym2)
print(" ")
print("Atome d'arrivée : Bi", BiSym)
print(" ")
print("Image :")

image_1=img.imread('data/Bi.png')
fig, axes = plt.subplots(figsize=(10,10))
axes.get_xaxis().set_visible(False)
axes.get_yaxis().set_visible(False)
axes.imshow(image_1)
plt.show()
```

## Troisième Symmétrie : 

```python
print("Troisième opération de symétrie")
print("------------------------------")
print(" ")
print("Atome de départ : Au", Au )
print(" ")
print("Opération de symétrie : Axe C2 : ")
print(Sym3)
print(" ")
print("Atome d'arrivée : Au", AuSym)
print(" ")
print("Image :")

image_1=img.imread('data/Au.png')
fig, axes = plt.subplots(figsize=(10,10))
axes.get_xaxis().set_visible(False)
axes.get_yaxis().set_visible(False)
axes.imshow(image_1)
plt.show()
```
