# Diffraction des rayons X sur un cristal Ba2BiAu

### Imports : 

```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
import matplotlib.pyplot as plt
```

### T√©l√©chargement du fichier CIF : 

```python
with MPRester("bfWHOpdayyP1LUqo") as m:

    # Structure for material id
    structure = m.get_structure_by_material_id("mp-862947")
```

```python
#Cr√©ation d'un objet XRDCalculator
xrd = XRDCalculator(1.54060)

#Calcul du mod√®le de diffraction
pattern = xrd.get_pattern(structure)
```

### Calculs des param√®tres des pics : 

Nous n'allons pas consid√©rer les pics 1 et 4 car leurs intensit√©s sont bien trop faibles compar√©es √† leurs voisins. Il est probable que ces pics proviennent d'interf√©rences parasites. 

```python
print('Intensit√© du premier pic : ' + str(pattern.y[0]))
print('Intensit√© du premier pic : ' + str(pattern.y[3]))
```

```python
#R√©cup√©ration de l'angle du deuxi√®me pic
theta2 = pattern.x[1]

#R√©cup√©ration de l'intensit√© du deuxi√®me pic
intensite2 = pattern.y[1]

#R√©cup√©ration des indices hkl du deuxi√®me pic
hkl2 = pattern.hkls[1]

#R√©cup√©ration de la distance inter-plan
d_hkls2 = pattern.d_hkls[1]

print('Angle du deuxi√®me pic : '+  '2ùúÉ = ' + str(theta2) + ' degr√©s')
print('Intensit√© du deuxi√®me pic : ' + str(intensite2))
print('Indices hkl du deuxi√®me pic :' + str(hkl2))
print("Valeur de la distance inter-plan :", d_hkls2)
```

```python
#R√©cup√©ration de l'angle du troisi√®me pic
theta3 = pattern.x[2]

#R√©cup√©ration de l'intensit√© du troisi√®me pic
intensite3 = pattern.y[2]

#R√©cup√©ration des indices hkl du troisi√®me pic
hkl3 = pattern.hkls[2]

#R√©cup√©ration de la distance inter-plan
d_hkls3 = pattern.d_hkls[2]

print('Angle du troisi√®me pic : '+  '2ùúÉ = '  + str(theta3) + ' degr√©s')
print('Intensit√© du troisi√®me pic : ' + str(intensite3))
print('Indices hkl du troisi√®me pic :' + str(hkl3))
print("Valeur de la distance inter-plan :", d_hkls3)
```

```python
#R√©cup√©ration de l'angle du troisi√®me pic
theta4 = pattern.x[4]

#R√©cup√©ration de l'intensit√© du troisi√®me pic
intensite4 = pattern.y[4]

#R√©cup√©ration des indices hkl du troisi√®me pic
hkl4 = pattern.hkls[4]

#R√©cup√©ration de la distance inter-plan
d_hkls4 = pattern.d_hkls[4]

print('Angle du troisi√®me pic : '+  '2ùúÉ = '  + str(theta4) + ' degr√©s')
print('Intensit√© du troisi√®me pic : ' + str(intensite4))
print('Indices hkl du troisi√®me pic :' + str(hkl4))
print("Valeur de la distance inter-plan :", d_hkls4)
```
